```python
def imprime_analise(result_path, text_path, ndocs):
    text_files = glob.glob(join(text_path, "*.txt"))
    result_files = glob.glob(join(result_path, "*.csv"))
    #for i in range(len(result_files)):
    for i in range(ndocs):
        print(result_files[-i-1])
        results = []
        with open(result_files[-i-1], encoding='utf-8') as file:
            for line in file:
                if not line.startswith(','):
                    results.append(line)
        with open(text_files[-i-1], encoding='utf-16') as file:
            texto = str(file.read())
            texto_colorido = ''
            texto_trecho = []
            start_texto = 0
            end_texto = len(texto)
            for result in results:
                result = result.replace('"PAF-3-MA,"', 'PAF-3-MA')
                result = result.replace('"Branca,2"', 'Branca')
                result = result.replace('"E,"', 'E')
                result = result.replace('"CP-1070,CP-1126,CP-1172"', 'CP-1070 e CP-1126 e CP-1172')
                result = result.replace('"TM-02,31,41D"', 'TM-02 31 41D')
                result = result.replace('"TM-03,16,26,49"', 'TM-03 16 26 49')
                result = result.replace('"RJS-,236"', '"RJS-236"')
                result = result.replace('"cretáciosQ,1e"', '"cretácios Q1e"')
                result = result.replace('"rocha,s"', '"rochas"')
                result = result.replace('"PB-18-RO,PB-29-ROe"', "PB-18-RO e PB-29-ROe")
                result = result.replace('","', '""')
                result = result.replace('"4-PA-2,1-CRO-l"', '"4-PA-2 1-CRO-l"')
                result = result.replace('"CaJ,."', '"CaJ."')
                result = result.replace('"I,"', '"I"')
                result = result.replace('"8S,8S8&8"', "8S8S8&8")
                result = result.replace('"Me-2-D,"', '"Me-2-D"')
                start_entity = int(result.strip('\n').split(',')[2])
                end_entity = int(result.strip('\n').split(',')[3])
                entity = result.strip('\n').split(',')[-1]
                texto_colorido = texto_colorido+cores[entity]
                texto_trecho.append(texto[start_texto:start_entity])
                texto_trecho.append(texto[start_entity:end_entity])
                start_texto = end_entity
            texto_colorido = texto_colorido+'\033[m {}'
            texto_trecho.append(texto[start_texto:end_texto])
        for k, v in cores.items():
            if k.startswith('B'):
                legenda = v.split(' ')[-2]+' {}'
                print(legenda.format(k))
        print(texto_colorido.format(*(t for t in texto_trecho)))                        
```

```python
imprime_analise(results_path, texts_path, 20)
```
