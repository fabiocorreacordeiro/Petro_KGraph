```python
from owlready2 import *
import pandas as pd
```

Carregando ontologia

```python
#OntoGeoLógica populada (OntoGeoLogicaPovoadaInstanciasRelacoes.owl)
onto_file = "../OntoGeoLogicaInstanciasRelacoes.owl"
onto = get_ontology(onto_file)
onto.load()
```

```python
# Dataframe para receber as triplas
triplas = pd.DataFrame(columns=['Termo', 'Related Term', 'Relation', 'Source'])
```

Classes e individuos

```python
classes = list(onto.classes())
individuals = list(onto.individuals())

print('Número de classes: ', len(classes))
print('Número de instâncias: ', len(list(individuals)))
```

```python
triplas
```

### Compilando os sinônimos

```python
# Compilando os sinônimos

n = 0
for conceito in classes + individuals:
    n = n + 1
    for label_i in range(len(conceito.label)):
        for label_j in range(label_i + 1, len(conceito.label)):
            triplas.loc[len(triplas.index)] = [conceito.label[label_i], conceito.label[label_j], 'SYN', onto_file]
            triplas.loc[len(triplas.index)] = [conceito.label[label_j], conceito.label[label_i], 'SYN', onto_file]
            #print(label_i, label_j)
    if n % 1000 == 0:
        print(n, ' conceitos processados')
    
#    if n > 20:
#        break
```

```python
triplas
```

### Escolhendo as relações que serão compiladas

```python
# Tordas as relações presentes na ontologia

for r in onto.object_properties():
    print (r)
```

```python
# Lista das relações que serão compiladas

relacoes = ['constituted_by', 'has_age', 'age_of', 'has_temporal_relation']
#'located_in', 'location_of','crosses', 'crossed_by',
```

```python
# iterando por todas as relações de interesse
for rel in relacoes:
    print('Iniciou relação: ', rel)
    
    # Selecionando a tupla de conceitos que possuem esse relação 
    n = 0
    for tupla_conceito in onto[rel].get_relations():
        n = n+ 1
        # Iterando pelos labels dos conceitos
        for label_1 in tupla_conceito[0].label:
            for label_2 in tupla_conceito[1].label:
                # Adicionando ao dataframe
                triplas.loc[len(triplas.index)] = [label_1, label_2, rel, onto_file]
                
        if n % 1000 == 0:
            print(n, ' relações ', rel, ' processadas')
                
    print('Finalizou relação: ', rel)
    print('------------')
        #break
```
