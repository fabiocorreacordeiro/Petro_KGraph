```python
# Creates file to store results
#with open("output.txt", "w+") as file:
def intrinsic_eval(model):
   
    pares_list = []
    likert = {}

    with open("likert_scale_intrinsic_test.txt") as file:
        for line in file:
            w = line.split(";")
            par = (w[0].rstrip(), w[1].rstrip())
            pares_list.append(par)
            likert.update({par:float(w[2].rstrip())})

    similaridadesDoModelo = []
    similaridadesDoLikert = []

    n = 0
    c = 0
    for par in pares_list:
        n = n + 1
        try:
            similarity = model.wv.similarity(par[0], par[1])
            similaridadesDoModelo.append(similarity)
            similaridadesDoLikert.append(likert.get((par[0], par[1])))
            #file.write(par[0]+"\t"+par[1]+"\t"+str(similarity)+"\n")
            
        except:
            c = c + 1
            #file.write(par[0]+"\t"+par[1]+"\t"+"NaN"+"\n")

    cSpearman = spearmanr(similaridadesDoModelo, similaridadesDoLikert)
    print("Spearman Correlation\t"+str(cSpearman[0]))
    print("Coverage\t" + str((n-c)/n))
    #file.write("Spearman Correlation\t"+str(cSpearman[0]))
```

```python
print('------------- \nPetroVec O&G público')
intrinsic_eval(Petrovec_OeG)
print('Número de Vetores: ', len(Petrovec_OeG.wv.vectors))
print('------------- \nPetroVec O&G NP2')
intrinsic_eval(Petrovec_OeG_NP2)
print('Número de Vetores: ', len(Petrovec_OeG_NP2.wv.vectors))
print('------------- \nPetroVec Híbrido público')
intrinsic_eval(Petrovec_hibrido)
print('Número de Vetores: ', len(Petrovec_hibrido.wv.vectors))
print('------------- \nPetroVec Híbrido NP2')
intrinsic_eval(Petrovec_hibrido_NP2)
print('Número de Vetores: ', len(Petrovec_hibrido_NP2.wv.vectors))

print('------------- \nPetroOntoVec_Base')
intrinsic_eval(PetroOntoVec_Base)
print('Número de Vetores: ', len(PetroOntoVec_Base.wv.vectors))
print('------------- \nPetroOntoVec no_onto_projection')
intrinsic_eval(PetroOntoVec_no_onto_projection)
print('Número de Vetores: ', len(PetroOntoVec_no_onto_projection.wv.vectors))
print('------------- \nPetroOntoVec com Petrovec_híbrido')
intrinsic_eval(PetroOntoVec_Petrovec_híbrido)
print('Número de Vetores: ', len(PetroOntoVec_Petrovec_híbrido.wv.vectors))
print('------------- \nPetroOntoVec com Petrovec_híbrido_NP2')
intrinsic_eval(PetroOntoVec_Petrovec_híbrido_NP2)
print('Número de Vetores: ', len(PetroOntoVec_Petrovec_híbrido_NP2.wv.vectors))
print('------------- \nPetroOntoVec com Petrovec_OeG')
intrinsic_eval(PetroOntoVec_Petrovec_OeG)
print('Número de Vetores: ', len(PetroOntoVec_Petrovec_OeG.wv.vectors))
print('------------- \nPetroOntoVec com Petrovec_OeG_NP2')
intrinsic_eval(PetroOntoVec_Petrovec_OeG_NP2)
print('Número de Vetores: ', len(PetroOntoVec_Petrovec_OeG_NP2.wv.vectors))
print('------------- \nPetroOntoVec Random_Walker')
intrinsic_eval(PetroOntoVec_Random_Walker)
print('Número de Vetores: ', len(PetroOntoVec_Random_Walker.wv.vectors))
print('------------- \nPetroOntoVec walk_depth_2')
intrinsic_eval(PetroOntoVec_walk_depth_2)
print('Número de Vetores: ', len(PetroOntoVec_walk_depth_2.wv.vectors))
print('------------- \nPetroOntoVec V2')
intrinsic_eval(PetroOntoVec_V2)
print('Número de Vetores: ', len(PetroOntoVec_V2.wv.vectors))
```
